# Feature Engineering

```python
testdata.isna().sum()
#we can see that the "fare" column has a null value. We fill it up.
```

```python
testdata = testdata.fillna({"Fare": 0})
```

# Training the Model

```python
from sklearn.model_selection import train_test_split
#importing train_test_split function as it is used to split a dataset into training and testing subsets.
```

```python
predictors = testdata[['Pclass', 'Sex', 'Parch']]
target = testdata["Survived"]
#in x variable we store all the factors that have a correlation with survival.
#in y variable we store the survival data
predictors
```

```python
x_train, x_test, y_train, y_test = train_test_split(predictors,target,test_size=0.2, random_state=0)
#dataset is split into 2 parts; training and testing
#80% is used for training and 20% used for testing
```

# Predicting survival using random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
#we will be using random forest classifier to train our model
#we will be using accuracy_score function to check how accurate our model is
```

```python
randomforest = RandomForestClassifier()
randomforest.fit(x_train, y_train)
y_pred = randomforest.predict(x_test)
y_pred
```

```python
# Finding the accuracy of the model
acc_randomforest = round(accuracy_score(y_pred, y_test) * 100, 2)
print(acc_randomforest)
```
